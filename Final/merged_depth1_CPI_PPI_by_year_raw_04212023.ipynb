```python
import pandas as pd
df = pd.read_csv('merged_1_depth_CPI_PPI.csv')
```

```python
df
```

```python
df['year_month'] = pd.to_datetime(df['year_month'], format='%Y/%m')
df['Year'] = df['year_month'].dt.year
mean_df = df.groupby('Year').mean()
merged_df = pd.merge(df['Year'].drop_duplicates(), mean_df, on='Year')
print(merged_df)
```

```python
merged_df.to_csv('merged_depth1_data_by_year_raw.csv', index=False)
```
