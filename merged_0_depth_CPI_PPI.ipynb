```python
import pandas as pd
import numpy as np
```

```python
#Read excel file FoodAccessResearchAtlasData2019, to see how many spreadsheet in that file
df_PPI = pd.read_csv('C:\\Users\\yueqi\\Yueqin_COMP5360_Final_Project\\Data\\BLS_Seafood_PPI.csv')
print(df_PPI)
df_CPI = pd.read_csv('C:\\Users\\yueqi\\Yueqin_COMP5360_Final_Project\\Data\\BLS_Seafood_CPI.csv')
print(df_CPI)
```

```python
# remove the "M" character from the period column
df_PPI["Period"] = df_PPI["Period"].str.replace("M", "")
df_CPI["Period"] = df_CPI["Period"].str.replace("M", "")
# print the updated dataframe
print(df_PPI)
print(df_CPI)
```

```python
# combine year and month columns in df_PPI and df_CPI
df_PPI['year_month'] = df_PPI['Year'].astype(str) + '/' + df_PPI['Period'].astype(str)
df_PPI
```

```python
#change column name "value" to "PPI"
df_PPI = df_PPI.rename(columns={'Value': 'PPI'})
df_PPI
```

```python
# combine year and month columns in df_PPI and df_CPI
df_CPI['year_month'] = df_CPI['Year'].astype(str) + '/' + df_CPI['Period'].astype(str)

df_CPI
```

```python
#change column name "value" to "CPI"
df_CPI = df_CPI.rename(columns={'Value': 'CPI'})
df_CPI
```

```python
# merge the dataframes on the "year_month" column
merged_CPI_PPI_df = pd.merge(df_CPI, df_PPI, on="year_month",how='outer')

# print the merged dataframe
print(merged_CPI_PPI_df)
```

```python
#Delete some columns that are not useful
merged_CPI_PPI_df =merged_CPI_PPI_df[["year_month", "Series ID_x", "CPI","Series ID_y" ,"PPI"]]
merged_CPI_PPI_df
```

```python
merged_CPI_PPI_df.to_csv('merged_CPI_PPI.csv', index=False)
```
