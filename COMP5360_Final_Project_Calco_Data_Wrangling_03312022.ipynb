Here's a brief description of each feature:

Cst_Cnt: Cast count (unique identifier for each cast, or deployment of sampling equipment)
Btl_Cnt: Bottle count (unique identifier for each water sample)
Sta_ID: Station ID (unique identifier for each sampling station)
Depth_ID: Depth ID (unique identifier for each depth at which a sample was collected)
Depthm: Depth of sample in meters
T_degC: Temperature in degrees Celsius
Salnty: Salinity (amount of salt in seawater)
O2ml_L: Dissolved oxygen concentration in milliliters per liter
STheta: Potential density relative to seawater (a measure of water mass density)
O2Sat: Oxygen saturation (percent of oxygen present in seawater compared to the maximum amount it can hold)
Oxy_µmol/Kg: Dissolved oxygen concentration in micromoles per kilogram
BtlNum: Bottle number (same as Btl_Cnt)
RecInd: Record indicator (indicates if the record is the best value for the parameter measured at that station and depth)
T_prec: Temperature precision
T_qual: Temperature quality flag
S_prec: Salinity precision
S_qual: Salinity quality flag
P_qual: Pressure quality flag
O_qual: Oxygen quality flag
SThtaq: Potential density quality flag
O2Satq: Oxygen saturation quality flag
ChlorA: Chlorophyll A concentration (a measure of phytoplankton biomass)
Chlqua: Chlorophyll A quality flag
Phaeop: Phaeophytin concentration (a measure of the breakdown of chlorophyll A)
Phaqua: Phaeophytin quality flag
PO4uM: Phosphate concentration in micromoles per liter
PO4q: Phosphate quality flag
SiO3uM: Silicate concentration in micromoles per liter
SiO3qu: Silicate quality flag
NO2uM: Nitrite concentration in micromoles per liter
NO2q: Nitrite quality flag
NO3uM: Nitrate concentration in micromoles per liter
NO3q: Nitrate quality flag
NH3uM: Ammonia concentration in micromoles per liter
NH3q: Ammonia quality flag
C14As1: Carbon-14 activity in the surface layer
C14A1p: Carbon-14 activity quality flag for the surface layer
C14A1q: Carbon-14 activity quality flag for the surface layer
C14As2: Carbon-14 activity in the deep layer
C14A2p: Carbon-14 activity quality flag for the deep layer
C14A2q: Carbon-14 activity quality flag for the deep layer
DarkAs: Dark bottle carbon-14 activity
DarkAp: Dark bottle carbon-14 activity quality flag
DarkAq: Dark bottle carbon-14 activity quality flag
MeanAs: Mean surface layer carbon-14 activity
MeanAp: Mean surface layer carbon-14 activity quality flag
MeanAq: Mean surface layer carbon-14 activity quality flag
IncTim: Incubation time for primary productivity measurement
LightP: Light percentage for primary productivity measurement
R_Depth: Reference depth
R_TEMP: Reference temperature
R_POTEMP: Reference potential temperature

```python
# extract year and month information from 'Depth_ID' column and create two new columns 'year' and 'month'
df['year'] = df['Depth_ID'].str[:2] + df['Depth_ID'].str[3:5]
df['month'] = df['Depth_ID'].str[5:7]

df
```

```python
df_grouped = df.groupby(['year', 'month']).mean().reset_index()
df_grouped
#So there are 376 rows as a total(time frame), some years not meansured monly
```

```python
#Since there are all different depthm used to measure the features, not sure depth will affect the value, and which depth should we use
# to explor the data, just ramdonly select one year to look into, here I use 19990
df_1990 = df[df['year'] == '1990']

# print the resulting dataframe
print(df_1990)
```

```python
#Try to see 1990 data the features that we are interested
df_1990_select = df_1990[['month','Depthm', 'T_degC','Salnty','pH2','pH1']]
df_1990_select
```

```python
#draw scatter plot of depthm and salnty, tempertaure, to see how depthm affect values
# extract data for March only
import pandas as pd
import matplotlib.pyplot as plt

df_march = df_1990_select[df_1990_select['month'] == '03']

# create a scatter plot of temperature vs depth
plt.scatter(df_march['Depthm'], df_march['T_degC'], label='Temperature', color='blue')
plt.scatter(df_march['Depthm'], df_march['Salnty'], label='Salinity', color='red')
plt.xlabel('Depth (m)')
plt.ylabel('Temperature (°C) / Salinity (psu)')
plt.title('Temperature and Salinity vs Depth in March')
plt.legend()
plt.show()

#So the salinity didint' vary much howeevr temperature changes, we still need to choose the same depth for all measurements to analysis
```
