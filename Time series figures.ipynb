```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_Salnty = df.dropna(subset=['Salnty'])
df_Salnty['Salnty'] = pd.to_numeric(df_Salnty['Salnty'], errors='coerce')
df_Salnty = df_Salnty[df_Salnty['Salnty'] > 30]
print(df_Salnty['Salnty'].dtype)
# set the color palette
#palette = sns.color_palette('hls', len(df['Year'].unique()))

# plot the time series
fig = sns.lineplot(x='Month', y='Salnty', data=df_Salnty, hue='Year')

# set the x-axis ticks and labels
fig.set_xticks(range(1, 13))
fig.set_xticklabels(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])

# display the plot
plt.show()
```

```python
fig=sns.lineplot(x='Year',y='Salnty',data=df_Salnty,hue='Month')
plt.show()
```
