```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('merged_depth1_calco_CPI_PPI_Major_species_Cleaned_04182023.csv')
df
```

```python
#Drop 1st, 2nd and 3rd column
df1 = df.drop(df.columns[[0,1,2]], axis=1)
df1
```

```python
#select only predictors and do correlation later
# Select columns 0 to 14
df_predictors = df1.iloc[:, 0:25]
df_predictors
#we have 25 predictors here, we will remove the ones closely correlated with the other one.
```

```python
# Calculate correlation matrix
corr_matrix = df_predictors.corr()

# Print correlation matrix
print(corr_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the heatmap
fig, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Show the plot
plt.show()
```

```python
#Based on correlation matrix and heatmap, we will delete some features that collinear related

predictors = df_predictors.columns
predictors
```

```python
#PPI and CPI are closely related 0.88, We will delete PPI since it has only 32 years of data, CPI has 58 years of data.
#O2ml_L Vs Oxy_ÃÂµmol/Kg  1
# Define the correlation threshold
#R_SIGMA, 'R_SVA', 'R_DYNHT' are closely related to 'T_degC', will delete them all but keep only 'T_degC'
#O2ml_L delete
#R_SIGMA	R_SVA	R_DYNHT	R_SIO3	R_PO4	R_NO3	R_NO2 will all deleted
#C14As1, C14As1 delete
#'R_CHLA'delete
#PPI delete

# Define a list of column names to drop
columns_to_drop = ['STheta','O2ml_L', 'R_SIGMA', 'R_SVA','R_DYNHT','R_SIO3','NO2uM','NO3uM','R_PO4','R_PHAEO','R_NO3','R_NO2','C14As1','C14As2','DarkAs','MeanAs','R_CHLA','PPI']

# Drop the specified columns
df_predictors_dropped = df_predictors.drop(columns=columns_to_drop)

# Print the resulting DataFrame
print(df_predictors_dropped)
```

```python
# Calculate correlation matrix after dropped those 
corr_matrix = df_predictors_dropped.corr()

# Print correlation matrix
print(df_predictors_dropped)
```

```python
# Calculate correlation matrix
corr_matrix2 = df_predictors_dropped.corr()

# Plot the heatmap
fig, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(corr_matrix2, annot=True, cmap='coolwarm')

# Show the plot
plt.show()
```
